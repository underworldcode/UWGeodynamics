# UWGeodynamics Installation Procedure


## Requirements

- Underworld2


## Pip install


The UWGeodynamics module can be installed directly from the Python Package Index:

```
    pip install UWGeodynamics
```

or

```
    pip install UWGeodynamics --user
```

## GitHub

The module is available through github at https://github.com/rbeucher/UWGeodynamics.git

```
    git clone https://github.com/rbeucher/UWGeodynamics.git
```

It can then be installed using

```
    pip install -e UWGeodynamics
```

We recommend using a local install of the package

```
    pip install -e UWGeodynamics --user
```

## Install on raijin


The module can be installed on raijin using the same procedure


```
    module load python/2.7.11
    git clone https://github.com/rbeucher/UWGeodynamics.git
    pip install -e UWGeodynamics --user
```

You will then need to make sure that you have set your PYTHONPATH properly.
The easiest is to add the following line in your pbs submission file:

```
   export PYTHONPATH=$PYTHONPATH:/path/to/UWGeodynamics
```

## Import the module


We recommand using:

```
   import UWGeodynamics as GEO
```
